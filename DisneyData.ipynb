```
# to find the total number of each genre (genre is a column header)

movies["genre"].value_counts()
```

```
# list top 3 genres

genre_count = movies["genre"].value_counts()
genre_count[:3]
```

```
# create a bar plot

genre_counts = movies["genre"].value_counts()
genre_counts.plot(kind="bar")
```

```
# find only the horror movies (horror is a type of genre & genre is a column header)

horror_movies = movies[movies["genre"] == "Horror"]
print(horror_movies)

print("________________________________________________________________________________________________________________")

black_comedy_movies = movies[movies["genre"] == "Black Comedy"]
print(black_comedy_movies)

```

```
# find the horror and black comedy movies

h_movies = movies["genre"] == "Horror"
bc_movies = movies["genre"] == "Black Comedy"

print(movies[h_movies | bc_movies])
```

```
# find all the comedy movies with total gross over 100 million 

comedy_movies = movies["genre"] == "Comedy"
min_gross = movies["total_gross"] > 100000000

print(movies[comedy_movies & min_gross])

# find number of comedy movies with total gross exceeding 100000000
print(movies[comedy_movies & min_gross].count())
```

```
# determine the number of comedy movies in each rating

is_comedy = movies["genre"] == "Comedy"
comedy_movies = movies[is_comedy]
c = comedy_movies["mpaa_rating"].value_counts()

# bar plot with ratings on x-axis, # of comedy movies on y-axis
c.plot(kind="bar")
```
